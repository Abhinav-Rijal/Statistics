```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Uniform distribution
f_x=[]
a=5
b=15
k=1/(b-a)
x=np.arange(a,b,0.1)
for i in x:
    f_x.append(k)
plt.plot(x,f_x)
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('Uniform probability density function')
```

```python
#Gaussian distribution
f_x=[]
mu=5
sigma=2.5
pi=3.14
k=1/(sigma*math.sqrt(2*pi))
x=np.arange(mu-5*sigma,mu+5*sigma,0.1)
for i in x:
    f=k*math.exp(-(((i-mu)/sigma)**2)/2)
    f_x.append(f)
plt.plot(x,f_x)
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('Gaussian probability density function')
```

```python
#Two Gaussian distributions comparison
f_x1=[]
f_x2=[]
mu=5
sigma1=2.5
sigma2=1
pi=3.14
k1=1/(sigma1*math.sqrt(2*pi))
x1=np.arange(mu-5*sigma1,mu+5*sigma1,0.1)
for i in x1:
    f1=k1*math.exp(-(((i-mu)/sigma1)**2)/2)
    f_x1.append(f1)
plt.plot(x1,f_x1,label='mu=5, sigma=2.5')

k2=1/(sigma2*math.sqrt(2*pi))
x2=np.arange(mu-5*sigma2,mu+5*sigma2,0.1)
for i in x2:
    f2=k2*math.exp(-(((i-mu)/sigma2)**2)/2)
    f_x2.append(f2)
plt.plot(x2,f_x2,label='mu=5, sigma=1')

plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('Gaussian probability density function')
plt.legend()
```

```python
#Cauchy distribution
f_x=[]
mu=5
lambd=10
pi=3.14
k=1/pi
x=np.arange(mu-5*lambd,mu+5*lambd,0.1)
for i in x:
    f=k*lambd/(lambd**2 + (i-mu)**2)
    f_x.append(f)
plt.plot(x,f_x)
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('Cauchy probability density function')
```

```python
#Gamma distribution; with two parameters alpha, beta
f_x=[]
alpha=2
beta=2
k=beta**alpha/math.gamma(alpha)
x=np.arange(0,10,0.1)
for i in x:
    f=k*math.exp(-beta*i)*i**(alpha-1)
    f_x.append(f)
plt.plot(x,f_x)
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('Gamma probability density function')
```

```python
def beta(m,n):
    return math.gamma(m)*math.gamma(n)/math.gamma(m+n)
```

```python
#Beta distribution of first kind
f_x=[]
m=2
n=2
k=1/beta(m,n)
x=np.arange(0,1.01,0.01)
for i in x:
    f=k*(i**(m-1))*(1-i)**(n-1)
    f_x.append(f)
plt.plot(x,f_x)
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('Beta probability density function of first kind')
```

```python
#Comparision of Beta distributions of first kind
f_x1=[]
m1=2
n1=4
k1=1/beta(m1,n1)
x=np.arange(0,1.01,0.01)
for i in x:
    f1=k1*(i**(m1-1))*(1-i)**(n1-1)
    f_x1.append(f1)
plt.plot(x,f_x1,label='m=2,n=4')

f_x2=[]
m2=2
n2=2
k2=1/beta(m2,n2)
x=np.arange(0,1.01,0.01)
for i in x:
    f2=k2*(i**(m2-1))*(1-i)**(n2-1)
    f_x2.append(f2)
plt.plot(x,f_x2,label='m=n=2')

f_x3=[]
m3=4
n3=2
k3=1/beta(m3,n3)
x=np.arange(0,1.01,0.01)
for i in x:
    f3=k3*(i**(m3-1))*(1-i)**(n3-1)
    f_x3.append(f3)
plt.plot(x,f_x3,label='m=4,n=2')

plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('Beta probability density function of first kind')
plt.legend()
```

```python
#Beta distribution of second kind
f_x=[]
m=3
n=4
k=1/beta(m,n)
x=np.arange(0,10.1,0.1)
for i in x:
    f=k*(i**(m-1))/(1+i)**(m+n)
    f_x.append(f)
plt.plot(x,f_x)
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('Beta probability density function of second kind')
```

```python
#Laplace distribution
f_x=[]
mu=5
theta=2

k=theta/2
x=np.arange(mu-5,mu+5,0.1) # 99% area within (-5,5)
for i in x:
    f=k*math.exp(-theta*abs(i-mu))
    f_x.append(f)
plt.plot(x,f_x)
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('Laplace probability density function')
```

```python
#Chi-squared distribution (from Gamma distribution)
n=5
f_x=[]
alpha=n/2
beta=1/2
k=beta**alpha/math.gamma(alpha)
x=np.arange(0,10*n,0.1)
for i in x:
    f=k*math.exp(-beta*i)*i**(alpha-1)
    f_x.append(f)
plt.plot(x,f_x)
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('Chi-squared probability density function')
```

```python
#Chi-squared distribution approaching Normal distribution for n>30
n=200
f_x=[]
alpha=n/2
beta=1/2
k=beta**alpha/math.gamma(alpha)
x=np.arange(n-3*math.sqrt(2*n),n+3*math.sqrt(2*n),0.1)
for i in x:
    f=k*math.exp(-beta*i)*i**(alpha-1)
    f_x.append(f)
plt.plot(x,f_x,label='Chi-squared')

f_x1=[]
mu=n
sigma=math.sqrt(2*n) #Chi-squared dist. approaches N(n,2n) for n>30
pi=3.14
k1=1/(sigma*math.sqrt(2*pi))
x=np.arange(mu-3*sigma,mu+3*sigma,0.1)
for i in x:
    f1=k1*math.exp(-(((i-mu)/sigma)**2)/2)
    f_x1.append(f1)
plt.plot(x,f_x1,label='normal')

plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('Chi-squared vs Normal probability density function')
plt.legend()
```

```python
#Student's t-distribution
n=11 #n>2
nu=n-1
f_x=[]

k=1/(math.sqrt(nu)*math.gamma(1/2)*math.gamma(nu/2)/math.gamma(1/2+nu/2))
print(math.gamma((1+nu)/2))
print(k)
x=np.arange(-3*math.sqrt(nu/(nu-2)),3*math.sqrt(nu/(nu-2))+0.1,0.1)
for i in x:
    f=k/(1+i**2/nu)**((1+nu)/2)
    f_x.append(f)
plt.plot(x,f_x)
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('Students t probability density function')
```

```python
#Student's t-distribution approaches normal distribution for large n
n=25
nu=n-1
f_x=[]

k=math.gamma(0.5+(nu/2))/(math.sqrt(nu)*math.gamma(1/2)*math.gamma(nu/2))
x=np.arange(-3*math.sqrt(nu/(nu-2)),3*math.sqrt(nu/(nu-2))+0.1,0.1)
for i in x:
    f=k*1/(1+i**2/nu)**((1+nu)/2)
    f_x.append(f)
plt.plot(x,f_x,label='Students t')

f_x1=[]
mu=0
sigma=math.sqrt(nu/(nu-2)) #Student's dist. approaches N(0,nu/(nu-2) for large n
pi=3.14
k1=1/(sigma*math.sqrt(2*pi))
x=np.arange(mu-3*sigma,mu+3*sigma,0.1)
for i in x:
    f1=k1*math.exp(-(((i-mu)/sigma)**2)/2)
    f_x1.append(f1)
plt.plot(x,f_x1,label='normal')
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('Students t vs Normal probability density function')
plt.legend()
```

```python
#F-distribution
m=5
n=5
f_x=[]
k=m/(n*beta(m/2,n/2))
x=np.arange(0.1,10.1,0.1)# x>0
for i in x:
    f=k*((m*i/n)**((m/2)-1))/(1+(m*i/n))**((m+n)/2)
    f_x.append(f)
plt.plot(x,f_x)
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('F-probability density function')
```
